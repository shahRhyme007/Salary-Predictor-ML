```python
import pandas as pd
import matplotlib.pyplot as plt

# reading the csv file
df = pd.read_csv("survey_results_public.csv")
```

```python
df.head()
```

```python
# cleaning the data frame

# the columns that we want to keep 
df = df[["Country", "EdLevel", "YearsCodePro", "Employment", "ConvertedCompYearly"]]

# renaming the component of the dframe
df = df.rename({"ConvertedCompYearly": "Salary"}, axis=1)
df.head()

```

```python

# we only need the data set where the salary is available
df = df [df['Salary'].notnull()]
df.head()
```

```python
df.info()
```

```python
# drop all the rows where one of those columns is not a number(CLEANING THE DATA)

df = df.dropna()
df.isnull().sum()
```

```python
# WE only want to keep the data where the user was employed fulltime
df = df[df["Employment"] == "Employed, full-time"]
df = df.drop("Employment", axis=1)
df.info()
```

```python
# cleaning the country data(no. of data points from each countries)
df['Country'].value_counts()
```

```python

# shortening the dsets where there is less data
def shorten_categories(categories, cutoff):
    categorical_map = {}
    for i in range(len(categories)):
        if categories.values[i] >= cutoff:
            categorical_map[categories.index[i]] = categories.index[i]
        else:
            # creating new catagory is less than cutoff
            categorical_map[categories.index[i]] = 'Other'
    return categorical_map
```

```python
country_map = shorten_categories(df.Country.value_counts(), 400)
df['Country'] = df['Country'].map(country_map)
df.Country.value_counts()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 7))
df.boxplot('Salary', 'Country', ax=ax)
plt.suptitle('Salary (US$) v Country')
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
df = df[df["Salary"] <= 300000]
df = df[df["Salary"] >= 10000]
df = df[df['Country'] != 'Other']
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 7))
df.boxplot('Salary', 'Country', ax=ax)
plt.suptitle('Salary (US$) v Country')
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
df["YearsCodePro"].unique()
```

```python
# converting everything to float
def clean_experience(x):
    if x == 'More than 50 years':
        return 50
    if x == 'Less than 1 year':
        return 0.5
    return float(x)
# applying the function 
df['YearsCodePro'] = df['YearsCodePro'].apply(clean_experience)
```

```python
# lets see the education array
df["EdLevel"].unique()
```

```python
# combining the degrees

def clean_education(x):
    if 'Bachelor’s degree' in x:
        return 'Bachelor’s degree'
    if 'Master’s degree' in x:
        return 'Master’s degree'
    if 'Professional degree' in x or 'Other doctoral' in x:
        return 'Post grad'
    return 'Less than a Bachelors'

df['EdLevel'] = df['EdLevel'].apply(clean_education)
```

```python
df["EdLevel"].unique()
```

```python
# converting the string into a unique number that the model can understand
from sklearn.preprocessing import LabelEncoder

le_education = LabelEncoder()
df['EdLevel'] = le_education.fit_transform(df['EdLevel'])
df["EdLevel"].unique()
```

```python
le_country = LabelEncoder()
df['Country'] = le_country.fit_transform(df['Country'])
df['Country'].unique()

```

```python
X = df.drop('Salary', axis=1)
y = df['Salary']
```

```python
# building the ML model
from sklearn.linear_model import LinearRegression

# creating the regression model
linear_reg = LinearRegression()
# fitting the data
linear_reg.fit(X, y.values)
```

```python
y_pred = linear_reg.predict(X)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

error = np.sqrt(mean_squared_error(y, y_pred))
```

```python
error
```

```python
# The error is very high. Lets try the DecisionTreeRegressor Model
```

```python
from sklearn.tree import DecisionTreeRegressor

dec_tree_reg = DecisionTreeRegressor(random_state=0)
# Fitting the decision tree regression model using the features X and the target variable y.
dec_tree_reg.fit(X, y.values)
```

```python
y_pred = dec_tree_reg.predict(X)
```

```python
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```

```python
# lets try again if we can make it better
```

```python
from sklearn.ensemble import RandomForestRegressor

random_forest_reg = RandomForestRegressor(random_state=0)
random_forest_reg.fit(X, y.values)
```

```python
y_pred = random_forest_reg.predict(X)

error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```

```python
# GridSearchCV giving us the best model for us
from sklearn.model_selection import GridSearchCV

max_depth = [None, 2, 4, 6, 8, 10, 12]
parameters = {"max_depth": max_depth}

regressor = DecisionTreeRegressor(random_state=0)
gs = GridSearchCV(regressor, parameters, scoring='neg_mean_squared_error')
gs.fit(X, y.values)
```

```python
regressor = gs.best_estimator_

regressor.fit(X, y.values)
y_pred = regressor.predict(X)
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```
