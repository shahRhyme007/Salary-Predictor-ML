```python
import pandas as pd
import matplotlib.pyplot as plt

# reading the csv file
df = pd.read_csv("survey_results_public.csv")
```

```python
df.head()
```

```python
# cleaning the data frame

# the columns that we want to keep 
df = df[["Country", "EdLevel", "YearsCodePro", "Employment", "ConvertedCompYearly"]]

# renaming the component of the dframe
df = df.rename({"ConvertedCompYearly": "Salary"}, axis=1)
df.head()

```

```python

# we only need the data set where the salary is available
df = df [df['Salary'].notnull()]
df.head()
```

```python
df.info()
```

```python
# drop all the rows where one of those columns is not a number(CLEANING THE DATA)

df = df.dropna()
df.isnull().sum()
```

```python
# WE only want to keep the data where the user was employed fulltime
df = df[df["Employment"] == "Employed, full-time"]
df = df.drop("Employment", axis=1)
df.info()
```

```python
# cleaning the country data(no. of data points from each countries)
df['Country'].value_counts()
```

```python

# shortening the dsets where there is less data
def shorten_categories(categories, cutoff):
    categorical_map = {}
    for i in range(len(categories)):
        if categories.values[i] >= cutoff:
            categorical_map[categories.index[i]] = categories.index[i]
        else:
            # creating new catagory is less than cutoff
            categorical_map[categories.index[i]] = 'Other'
    return categorical_map
```

```python
country_map = shorten_categories(df.Country.value_counts(), 400)
df['Country'] = df['Country'].map(country_map)
df.Country.value_counts()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 7))
df.boxplot('Salary', 'Country', ax=ax)
plt.suptitle('Salary (US$) v Country')
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
df = df[df["Salary"] <= 300000]
df = df[df["Salary"] >= 10000]
df = df[df['Country'] != 'Other']
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 7))
df.boxplot('Salary', 'Country', ax=ax)
plt.suptitle('Salary (US$) v Country')
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```
